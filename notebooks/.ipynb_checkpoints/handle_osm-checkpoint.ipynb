```python
import pandas as pd
from io import open
import json as json
import re
```

```python
archivo = open("../osm/edificio_este.osm","r")

osm = archivo.readlines()
```

```python
def load_object(osm,OSM_OBJETO):
    # identifica donde inicia un objeto y donde 
    # termina, y los pone en una lista    
    objeto = []
    coma   = []
    for i,renglon in enumerate(osm):
        if renglon.startswith("OS:"):
            objeto.append(i)
        if ";" in renglon:
            coma.append(i)
#     en os_lista, guarda donde inicia y termina cada objeto
    osm_lista = []
    for i,j in zip(objeto,coma):
        osm_lista.append(osm[i:j])
# separa en dos listas, lista objeto y resto del osm
    objeto_lista = []
    resto_lista  = []
    for objeto in osm_lista:
        if OSM_OBJETO in objeto:
            objeto_lista.append(objeto)
        else:
            resto_lista.append(objeto)
#             Limpia la lista que contiene el objeto
    lista_materiales = []
    for material in objeto_lista:
        lista_material = []
        material[0]= material[0].replace(",","")
        for campo in material:
            lista_material.append(campo.replace("\n","").
                                 replace("","").
                                 replace("{","").
                                 replace("}","").
                                 replace("!- ",""))
#             .replace("{",""))
#                                   .replace("}","")
#                                   .replace("!- ","").replace("'","")
#                                   .replace("\n",""))
        lista_materiales.append(lista_material)
            
    return lista_materiales,resto_lista
materiales, restante = load_object(osm,"OS:Material,\n")
materiales
```
